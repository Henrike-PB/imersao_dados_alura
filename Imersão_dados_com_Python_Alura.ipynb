```python
# Traduzindo os nomes das colunas para português
coluna_pt_br = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df.rename(columns=coluna_pt_br, inplace=True)

# Exibindo as colunas traduzidas
df.columns
```

```python
# Traduzindo as categorias da coluna 'senioridade'
traducao_senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(traducao_senioridade)

# Exibindo as categorias traduzidas para verificar
df['senioridade'].value_counts()
```

```python
# Traduzindo as categorias da coluna 'contrato'
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'].replace(traducao_contrato, inplace=True)

# Exibindo as categorias traduzidas para verificar
print(df['contrato'].value_counts())
```

```python
# Traduzindo as categorias da coluna 'tamanho_empresa'
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'].replace(traducao_tamanho_empresa, inplace=True)

# Exibindo as categorias traduzidas para verificar
df['tamanho_empresa'].value_counts()
```

```python
# Traduzindo as categorias da coluna 'remoto'
traducao_remoto = {
    100: 'Remoto',
    0: 'Presencial',
    50: 'Híbrido'
}

df['remoto'] = df['remoto'].replace(traducao_remoto)

# Exibindo as categorias traduzidas para verificar
df['remoto'].value_counts()
```

```python
import numpy as np


#Criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a média salarial e substitui os nulos pela média e arrendonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))


'''Calcula mediana e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

#DESAFIO:
##Gerar um gráfico com plotly da média salarial de Cientista de Dados por país

```python
'''Primeiro devemos filtrar o dataframe com apenas os dados que queremos trabalhar, neste caso será o cargo de Cientista de Dados, ou seja selecionamos
   todas as linhas de 'df_limpo' onde a coluna 'cargo' é igual a 'Data Scientist'
'''

df_cientista_dados = df_limpo[df_limpo['cargo'] == 'Data Scientist']

cientista_dados_salario_medio_pais = df_cientista_dados.groupby('empresa')['usd'].mean().sort_values(ascending=False).reset_index()
fig_media_salario_pais = px.bar(cientista_dados_salario_medio_pais,
                                x='empresa',
                                y='usd',
                                title='Media salarial de Cientista de Dados por País',
                                labels={'empresa': 'País', 'usd': 'Média Salarial Anual (USD)'}
                                )
fig_media_salario_pais.show()
```
